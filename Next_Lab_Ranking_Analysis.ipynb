```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("browser_rankings_data.csv" ,encoding='ISO-8859-1' )
df.shape
```

```python
df = df.dropna()
df.head(20)
```

```python
df.describe()
```

## Let's group by rank and APP ID by average ranking of APP ID , So we can see that which app got better overall ranking
### Lower the rank better it is

```python
df_APP = df.groupby(["App ID"])[["Rank"]].mean()
```

### We can see that com.brave.browser	com.cloudmosa.puffinTV	com.opera.browser	net.fast.web.browser	has good ranking 
### and we can also see the average ranking of others 

```python
df_APP
```

## NOW Let's do group by using multiple columns 
## Like in this one we can see that for each keyword and for that Keyword how APPID ranking is given

```python
df_GP = df.groupby(["Keyword" , "App ID"])[["Rank"]].mean()
df_GP
```

## Now for code below we can see that how for same short description if Keyword changes then how with same short description rank changes
